# Código para Criação de Matriz par a par
---

Adaptado a partir do artigo: [Deep dive into Analytical Hierarchy Process using Python (Anirudh Chandra)](https://archive.is/6PLXj)

```python
# Importar Numpy
import numpy as np

# Criar Variável [Critérios]
criterios = ['Critério1','Critério2','Critério3','Critério4']
        
# Cria matriz padrão com números 1
n = len(criterios) # extrai quantidade de Critérios
A = np.ones([n,n]) # cria matriz populada com números 1

# Cria um loop para coletar os inputs ds usuário e popular o triângulo superior da matriz
for i in range(0,n):
    for j in range(0,n):
         if i<j:
             opi = criterios[i]
             opj = criterios[j]
             aij = input('Qual a importância de {} em relação a {} ?: '.format(opi,opj))
             A[i,j] = float(aij)   # triângulo superior
             A[j,i] = 1/float(aij) # triângulo inferior

# Mostrar Matriz Gerada:
print('---------------------')
print('Matriz de Comparação Par a Par:')
print(A)
```

```python
# Extra: Plot
import seaborn as sns
sns.heatmap(A, annot = True, xticklabels=criterios, yticklabels=criterios, cmap="Spectral")
```
